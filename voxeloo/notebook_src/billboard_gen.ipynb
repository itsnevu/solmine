```python
def to_rgba(array):
    r = array[:, :, :, 0].astype(np.uint32) << 24
    g = array[:, :, :, 1].astype(np.uint32) << 16
    b = array[:, :, :, 2].astype(np.uint32) << 8
    a = array[:, :, :, 3].astype(np.uint32)
    return r | g | b | a
```

```python
grass_image = Image.open(f"../../../voxeloo-web/public/textures/red_tulip.png")
```

```python
voxels = np.zeros(shape=(16, 16, 16, 4), dtype=np.uint32)
voxels[:, :, 8, :] = grass_image
```

```python
visualize_3d_window(
    voxels_to_mesh(to_rgba(voxels))
)
```
