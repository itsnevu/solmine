```python
import numpy as np
from pyvox.models import Vox
from pyvox.parser import VoxParser
from pyvox.writer import VoxWriter
```

```python
def to_rgba(array):
    r = array[:, :, :, 0].astype(np.uint32) << 24
    g = array[:, :, :, 1].astype(np.uint32) << 16
    b = array[:, :, :, 2].astype(np.uint32) << 8
    a = array[:, :, :, 3].astype(np.uint32)
    return r | g | b | a

def save_img(pixels, name):
    Image.fromarray(pixels).save(
        f"C:/Users/Taylor/Code/voxeloo-web/public/textures/{name}.png"
    )
```

```python
def extract_textures(input_path, output_folder):
    input_path = Path(input_path)
    output_folder = Path(output_folder)
    
    output_folder.mkdir(parents=True, exist_ok=True)
    voxels = VoxParser(input_path).parse().to_dense_rgba()
    
    top = Image.fromarray(voxels[:, 0, :])
    bottom = Image.fromarray(voxels[:, -1, :])
    side = Image.fromarray(voxels[0, :, :])
    
    output_base = input_path.with_suffix("").stem
    
    top.save(output_folder / f"{output_base}_top.png")
    bottom.save(output_folder / f"{output_base}_bottom.png")
    side.save(output_folder / f"{output_base}_side.png")
```
