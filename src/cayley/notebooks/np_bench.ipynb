```python
import numpy as np
```

```python
a = np.full([100, 100], 1)
b = np.full([100, 100], 2)
c = np.full([100, 100], 3)
```

```python
x = np.full([100, 100], 0)
```

```python
%%timeit

x[...] = a
```

```python
%%timeit

x[...] = a + b
```

```python
%%timeit

x[...] = a + b + c
```

```python
%%timeit

x[1:-1, 1:-1] = a[1:-1, 1:-1]
```

```python
%%timeit

x[1:-1, 1:-1] = a[1:-1, 1:-1] + b[1:-1, 1:-1]
```

```python
%%timeit

x[1:-1:, 1:-1] = a[1:-1, 1:-1] + b[1:-1, 1:-1] + c[1:-1, 1:-1]
```

```python
%%timeit

voxels = np.zeros(shape=[32, 32, 32], dtype=np.uint32)
voxels[:32,:1,:32] = np.ones([32, 1, 32])
voxels[14:18,1:32,14:18] = 2 * np.ones([4, 31, 4])
voxels[...] += voxels
```
