```python
import json
from glob import glob
```

### Read in all of the blocks

```python
blocks = {}
for path in glob("../../data/blocks/*.json"):
    with open(path) as f:
        blocks[path] = json.load(f)
```

### Do some transformations

```python
for path, block in blocks.items():
    if "label" in block:
        del block["label"]
    if "index" in block:
        del block["index"]
    for sample in block["samples"]:
        if sample["criteria"]["dye"] == "empty":
            sample["criteria"]["dye"] = "none"
```

### Write out all of the blocks

```python
for path, block in blocks.items():
    with open(path, "w") as f:
        json.dump(block, f, indent=2)
```
